```python
import pandas as pd
```

```python
# List of CSV files to merge
files = ["Sales_January_2019.csv","Sales_February_2019.csv","Sales_March_2019.csv","Sales_April_2019.csv","Sales_May_2019.csv",
         "Sales_June_2019.csv","Sales_July_2019.csv","Sales_August_2019.csv",
         "Sales_September_2019.csv","Sales_October_2019.csv","Sales_November_2019.csv","Sales_December_2019.csv"]

# Read and concatenate the CSV files
df = [pd.read_csv(file) for file in files]
data = pd.concat(df, ignore_index=True)

# Write the merged DataFrame to a new CSV file
data.to_csv('merged_file.csv', index=False)
```

```python
data = data[data['Order Date'].str[0:2] != 'Or']
data.head()
```

```python
data["Month"] = data["Order Date"].str[0:2]
data['Month'] = data['Month'].astype('int32')
data.head()
```

```python
data["Quantity Ordered"] = pd.to_numeric(data["Quantity Ordered"])
data["Price Each"] = pd.to_numeric(data["Price Each"])
```

```python
data["Sales"] = data["Quantity Ordered"] * data["Price Each"]
data.head()
```

```python
def get_city(address):
    return address.split(',')[1]
def get_state(address):
    return address.split(',')[2].split(' ')[1]
data["City"] = data["Purchase Address"].apply(lambda x: get_city(x) + ' (' + get_state(x)+ ')')
data.head()
```

```python
total_sale = data.groupby("Month").sum()
total_sale
```

```python
from matplotlib import pyplot as plt 
months = range(1,13)
plt.bar(months,total_sale["Sales"])
plt.xticks(months)
plt.ylabel("Sale in USD($)")
plt.xlabel('Month')
plt.show()
```

```python
total_sale = data.groupby('City').sum()
total_sale
```

```python
cities = [city for city, df in data.groupby('City')]
plt.bar(cities,total_sale["Sales"])
plt.xticks(cities,rotation='vertical',size=7)
plt.ylabel("Sale in USD($)")
plt.xlabel('Cities')
plt.show()
```

```python
hour = [hour for hour, df in data.groupby('Hour')]
plt.plot(hour,data.groupby(['Hour']).count())
plt.xticks(hour)
plt.xlabel('Hours')
plt.ylabel('Number of orders')
plt.grid()
plt.show()
```

```python
df = data[data['Order ID'].duplicated(keep=False)]
df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
df = df[['Order ID','Grouped']].drop_duplicates()
df.head()
```

```python
from itertools import combinations
from collections import Counter
count = Counter()
for row in df['Grouped']:
    rowlist = row.split(',')
    count.update(Counter(combinations(rowlist,2)))
    
for key, value in count.most_common(10):
    print(key,value)
```

```python
product_grouped = data.groupby('Product')
quantity_ordered = product_grouped.sum()['Quantity Ordered']

products = [product for product,df in product_grouped]
plt.bar(products,quantity_ordered)
plt.ylabel('Quantity Ordered')
plt.xlabel('Products')
plt.xticks(products,rotation='vertical',size=7)
plt.show()
```

```python
price = data.groupby('Product').mean()['Price Each']


fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(products,quantity_ordered,color='g')
ax2.plot(products, price, 'b-')

ax1.set_xlabel('Products Name')
ax1.set_ylabel('Quantity Ordered', color='g')
ax2.set_ylabel('Price ($)', color='b')
ax1.set_xticklabels(products,rotation='vertical',size=7)
plt.show()
```
